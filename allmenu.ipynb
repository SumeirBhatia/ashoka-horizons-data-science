```python
data = {
    'Age': [10,20,25,40,50,60,79,80,90],
    'weight':[15,25,35,55,25,45,35,55, 30]
}
data = pd.DataFrame(data)
```

```python
plt.plot(data['Age'], data['weight'])
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Age'], data['weight'], marker='o', color='blue', markeredgewidth='3.0', markeredgecolor='black', markerfacecolor='black')
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Age'], data['weight'], marker='v', color='green', markersize='10.0', markerfacecolor='purple', linestyle='-.')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight Line Plot', fontsize=35, color='red')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(data['Age'], data['weight'], marker='o', color='red', markersize='10.0', markerfacecolor='blue', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight Line Plot', fontsize=20, color='magenta')

plt.subplot(1,2,2)
plt.bar(x=df['cat'].unique(), height=df['cat'].value_counts().to_list(), color=['#0f5a3e', '#f8b195', '#f67280'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of Categories', fontsize=30, color='green')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Age'], data['weight'], marker='o', color='black', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.fill_between(data['Age'], data['weight'], color='cyan')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight Line Plot', fontsize=20, color='green')
plt.show()
```
